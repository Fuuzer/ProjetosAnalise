```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import calendar
```

```python
data = pd.read_csv("dados/Arquivo_NetFlix.csv", sep=',')
data
```

```python
# Verifique se existem valores nulos na coluna "date_added"
null_count = data['date_added'].isnull().sum()

# Exiba o nÃºmero de valores nulos
print(f"Total de valores nulos na coluna 'date_added': {null_count}")
```

```python
data = data.dropna(subset=['date_added'])
```

```python
total_filmes = data[data['type'] == 'Movie'].shape[0]
total_tv_shows = data[data['type'] == 'TV Show'].shape[0]
total_geral = total_filmes + total_tv_shows

print("Total de Filmes:", total_filmes)
print("Total de TV Shows:", total_tv_shows)
print("Total Geral:", total_geral)
```

```python
plt.style.use('dark_background')

dados = [total_filmes, total_tv_shows]

# RÃ³tulos para as fatias da pizza
rotulos = ['Filmes', 'TV Shows']

# Cores para as fatias
dados = [total_filmes, total_tv_shows]

# RÃ³tulos para as fatias da pizza
rotulos = ['Filmes', 'TV Shows']

# Cores atraentes
cores = ['#FF9999', '#66B3FF']

# Explode para destacar uma fatia (opcional)
explode = (0.1, 0)

# Crie o grÃ¡fico de pizza
plt.figure(figsize=(10, 6))  # Ajuste o tamanho da figura
plt.pie(dados, labels=rotulos, colors=cores, autopct='%1.1f%%', startangle=140, explode=explode)

# TÃ­tulo do grÃ¡fico
plt.title("ProporÃ§Ã£o entre Filmes e TV Shows")

# Legenda
plt.legend()

# Mostre o grÃ¡fico
plt.axis('equal')  # Garante que o grÃ¡fico seja uma pizza perfeita
plt.show()
```

```python
# Solicitar ao usuÃ¡rio que insira o ano desejado
ano_desejado = input("Digite o ano desejado (de 2008 a 2021): ")

# Converter a entrada do usuÃ¡rio para um nÃºmero inteiro
ano_desejado = int(ano_desejado)

# Filtrar os dados para o ano desejado
dados_ano_desejado = data[data['release_year'] == ano_desejado]

# Contar a quantidade de filmes e TV shows para o ano selecionado
total_filmes = dados_ano_desejado[dados_ano_desejado['type'] == 'Movie'].shape[0]
total_tv_shows = dados_ano_desejado[dados_ano_desejado['type'] == 'TV Show'].shape[0]

dados = [total_filmes, total_tv_shows]

# RÃ³tulos para as fatias da pizza
rotulos = ['Filmes', 'TV Shows']

# Cores atraentes
cores = ['#FF9999', '#66B3FF']

# Explode para destacar uma fatia (opcional)
explode = (0.1, 0)

# Crie o grÃ¡fico de pizza
plt.figure(figsize=(10, 6))  # Ajuste o tamanho da figura
plt.pie(dados, labels=rotulos, colors=cores, autopct='%1.1f%%', startangle=140, explode=explode)

# TÃ­tulo do grÃ¡fico
plt.title(f"ProporÃ§Ã£o entre Filmes e TV Shows em {ano_desejado}")

# Legenda
plt.legend()

# Mostre o grÃ¡fico
plt.axis('equal')  # Garante que o grÃ¡fico seja uma pizza perfeita
plt.show()
```

```python
# Trate a coluna "country" para dividir mÃºltiplos paÃ­ses em entradas separadas
data['country'] = data['country'].str.split(', ')
data = data.explode('country')

# Calcule as contagens de paÃ­ses para o total, filmes e TV shows
total_countries = data['country'].value_counts().head(10)
movie_countries = data[data['type'] == 'Movie']['country'].value_counts().head(10)
tv_show_countries = data[data['type'] == 'TV Show']['country'].value_counts().head(10)

# Crie grÃ¡ficos de barra para os 10 principais paÃ­ses em cada categoria
plt.figure(figsize=(20, 6))

# Ajuste o espaÃ§o entre as barras
bar_width = 0.5

# GrÃ¡fico para o total
plt.subplot(131)
total_countries.plot(kind='bar', color='skyblue', width=bar_width)
plt.title('Top 10 PaÃ­ses (Total)')
plt.xticks(rotation=45)

# GrÃ¡fico para filmes
plt.subplot(132)
movie_countries.plot(kind='bar', color='lightcoral', width=bar_width)
plt.title('Top 10 PaÃ­ses (Filmes)')
plt.xticks(rotation=45)

# GrÃ¡fico para TV shows
plt.subplot(133)
tv_show_countries.plot(kind='bar', color='lightgreen', width=bar_width)
plt.title('Top 10 PaÃ­ses (TV Shows)')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Separe os diretores mÃºltiplos em entradas separadas
data['director'] = data['director'].str.split(', ')
data = data.explode('director')

# Calcule as contagens de diretores para o total, filmes e TV shows
total_directors = data['director'].value_counts().head(10)
movie_directors = data[data['type'] == 'Movie']['director'].value_counts().head(10)
tv_show_directors = data[data['type'] == 'TV Show']['director'].value_counts().head(10)

# Crie grÃ¡ficos de barra para os 10 diretores mais frequentes em cada categoria
plt.figure(figsize=(25, 7))

# FunÃ§Ã£o para adicionar nÃºmeros Ã s barras
def add_bar_numbers(ax, data_series):
    for i, v in enumerate(data_series):
        ax.text(i, v, str(v), ha='center', va='bottom')

# GrÃ¡fico para o total
plt.subplot(131)
total_directors.plot(kind='bar', color='skyblue')
plt.title('Top 10 Diretores (Total)')
plt.xticks(rotation=45)
add_bar_numbers(plt.gca(), total_directors)

# GrÃ¡fico para filmes
plt.subplot(132)
movie_directors.plot(kind='bar', color='lightcoral')
plt.title('Top 10 Diretores (Filmes)')
plt.xticks(rotation=45)
add_bar_numbers(plt.gca(), movie_directors)

# GrÃ¡fico para TV shows
plt.subplot(133)
tv_show_directors.plot(kind='bar', color='lightgreen')
plt.title('Top 10 Diretores (TV Shows)')
plt.xticks(rotation=45)
add_bar_numbers(plt.gca(), tv_show_directors)

plt.tight_layout()
plt.show()
```

```python
# Extrai o ano da coluna "date_added" e cria uma nova coluna "year_added"
data['year_added'] = data['date_added'].str[-4:]

# Conta a quantidade de tÃ­tulos adicionados em cada ano
year_counts = data['year_added'].value_counts().sort_index()

# Crie um grÃ¡fico de barras para visualizar os anos e a quantidade de tÃ­tulos adicionados
plt.figure(figsize=(12, 6))
year_counts.plot(kind='bar', color='skyblue')
plt.title('Quantidade de TÃ­tulos Adicionados por Ano')
plt.xlabel('Ano de AdiÃ§Ã£o')
plt.ylabel('Quantidade de TÃ­tulos')
plt.xticks(rotation=45)

# Adicione os nÃºmeros correspondentes a cada barra
for i, count in enumerate(year_counts):
    plt.text(i, count, str(count), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
# Filtra os tÃ­tulos do tipo "Movie"
movie_data = data[data['type'] == 'Movie']

# Conta a quantidade de filmes adicionados em cada ano
movie_year_counts = movie_data['year_added'].value_counts().sort_index()

# Crie um grÃ¡fico de barras para visualizar os anos e a quantidade de filmes adicionados
plt.figure(figsize=(12, 6))
movie_year_counts.plot(kind='bar', color='lightcoral')
plt.title('Quantidade de Filmes (Movies) Adicionados por Ano')
plt.xlabel('Ano de AdiÃ§Ã£o')
plt.ylabel('Quantidade de Filmes (Movies)')
plt.xticks(rotation=45)

# Adicione os nÃºmeros correspondentes a cada barra
for i, count in enumerate(movie_year_counts):
    plt.text(i, count, str(count), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
tv_show_data = data[data['type'] == 'TV Show']

# Conta a quantidade de sÃ©ries (TV Shows) adicionados em cada ano
tv_show_year_counts = tv_show_data['year_added'].value_counts().sort_index()

# Crie um grÃ¡fico de barras para visualizar os anos e a quantidade de sÃ©ries (TV Shows) adicionadas
plt.figure(figsize=(12, 6))
tv_show_year_counts.plot(kind='bar', color='lightgreen')
plt.title('Quantidade de SÃ©ries (TV Shows) Adicionadas por Ano')
plt.xlabel('Ano de AdiÃ§Ã£o')
plt.ylabel('Quantidade de SÃ©ries (TV Shows)')
plt.xticks(rotation=45)

# Adicione os nÃºmeros correspondentes a cada barra
for i, count in enumerate(tv_show_year_counts):
    plt.text(i, count, str(count), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
# Remova espaÃ§os extras no inÃ­cio das datas na coluna "date_added"
data['date_added'] = data['date_added'].str.strip()

# Converta a coluna "date_added" para o formato de data
data['date_added'] = pd.to_datetime(data['date_added'], format='%B %d, %Y', errors='coerce')

# Crie uma nova coluna "month_added" com o mÃªs extraÃ­do
data['month_added'] = data['date_added'].dt.strftime('%B')
```

```python
# Solicitar ao usuÃ¡rio que insira o ano desejado
ano_desejado = input("Digite o ano desejado (de 2008 a 2021): ")

# Converter a entrada do usuÃ¡rio para um nÃºmero inteiro
ano_desejado = int(ano_desejado)

# Filtrar os dados para o ano desejado
dados_ano_desejado = data[data['release_year'] == ano_desejado]

# Criar uma lista com os nomes dos meses em ordem
meses_em_ordem = [calendar.month_name[i] for i in range(1, 13)]

# Contar a quantidade de tÃ­tulos por mÃªs separados por tipo (Movie e TV Show)
contagem_por_mes = dados_ano_desejado.groupby(['month_added', 'type']).size().unstack(fill_value=0)

# Reordenar o DataFrame de acordo com os meses em ordem
contagem_por_mes = contagem_por_mes.reindex(meses_em_ordem)

# Configurar os dados para o grÃ¡fico
meses = contagem_por_mes.index
quantidades_filmes = contagem_por_mes['Movie']
quantidades_series = contagem_por_mes['TV Show']

# Criar o grÃ¡fico de barras
plt.figure(figsize=(12, 6))
bar_width = 0.35
bar_positions = range(len(meses))

plt.bar(bar_positions, quantidades_filmes, bar_width, color='skyblue', label='Filmes')
plt.bar([pos + bar_width for pos in bar_positions], quantidades_series, bar_width, color='lightcoral', label='SÃ©ries')

plt.title(f'Quantidade de TÃ­tulos Adicionados por MÃªs em {ano_desejado}')
plt.xlabel('MÃªs')
plt.ylabel('Quantidade de TÃ­tulos')
plt.xticks([pos + bar_width / 2 for pos in bar_positions], meses, rotation=45)

plt.legend(loc='upper right')

# Adicionar os nÃºmeros correspondentes a cada barra
for i, count in enumerate(bar_positions):
    plt.text(count, quantidades_filmes[i], str(quantidades_filmes[i]), ha='center', va='bottom')
    plt.text(count + bar_width, quantidades_series[i], str(quantidades_series[i]), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
contagem_por_mes = data['month_added'].value_counts().reindex(calendar.month_name[1:], fill_value=0)

# Configurar os dados para o grÃ¡fico
quantidades = contagem_por_mes.values

# Criar o grÃ¡fico de barras
plt.figure(figsize=(12, 6))
plt.bar(meses, quantidades, color='skyblue')
plt.title('Quantidade Total de TÃ­tulos Adicionados por MÃªs')
plt.xlabel('MÃªs')
plt.ylabel('Quantidade de TÃ­tulos')
plt.xticks(rotation=45)

# Adicionar os nÃºmeros correspondentes a cada barra
for i, count in enumerate(quantidades):
    plt.text(i, count, str(count), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
